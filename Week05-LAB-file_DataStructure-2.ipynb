### Activity 1: 

Write a program that:
1. Creates a file called "my_profile.txt"
2. Writes 3 lines: your name, your age, and your favorite color
3. Reads the file back and prints each line with a label

**Example output:**
"Name: Alex", "Age: 20", "Favorite Color: Blue"
<hr>

`````python
# define the file
def write_profile():
    with open("my_profile.txt", "w") as file:
        file.write("Shamsi\n")        
        file.write("20\n")             
        file.write("Pink")          

def read_profile():
    with open("my_profile.txt", "r") as file:
        lines = file.readlines()
        print(f'Name: {lines[0].strip()}')
        print(f'Age: {lines[1].strip()}')
        print(f'Favorite Color: {lines[2].strip()}')
# main function
def main():
    write_profile()
    read_profile()

if __name__ == "__main__":
    main()
`````

### Activity 2: Simple Shopping List

Create a program that:
1. Writes a shopping list to "shopping.txt" (5 items, one per line)
2. Reads the file and prints "You need to buy: [item]" for each item
3. Counts and displays the total number of items

<hr>

`````python
# define the file 
def write_shopping_list():
    items = ["vegetables",      # 5 items in a list 
             "spices",
             "milk", 
             "beans", 
             "toiletpaper",
            ]
    with open("shopping.txt", "w") as file:
        for item in items:
            file.write(item + "\n")

def read_and_display_list():
    try:
        with open("shopping.txt", "r") as file: #open file in read mode "r"
            lines = file.readlines()
            print(" Shopping List")
            for item in lines:
                print(f"You need to buy: {item.strip()}")
            print(f"Total items: {len(lines)}")
    except FileNotFoundError:
        print("Error: 'shopping.txt' not found.")

def main():
    write_shopping_list()
    read_and_display_list()

if __name__ == "__main__":
    main()
`````

### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
# define file
def write_temperatures():
    temperatures = []
    with open("temps.txt", "w") as file:
        for day in range(1, 6):
            while True:
                try:        #User prompt
                    temp = float(input(f"Enter temperature for Day {day} (°C): "))
                    temperatures.append(temp)
                    file.write(f"Day {day}: {temp}°C\n")
                    break
                except ValueError:
                    print("Please enter a valid number.")
    return temperatures

def read_and_display_average(temperatures):
    with open("temps.txt", "r") as file:
        print("\nRecorded Temperatures:")
        for line in file:
            print(line)
    average = sum(temperatures) / len(temperatures)
    print(f"\nAverage Temperature: {average:.2f}°C")

def main():
    temps = write_temperatures()
    read_and_display_average(temps)

if __name__ == "__main__":
    main()
`````

---

`````python
# Movie collection tuple 
movies = (("Inception", 2010), ("Matrix", 1999), ("Interstellar", 2014))

# Print the oldest movie
oldest_movie = min(movies, key=lambda x: x[1])
print("The oldest movie:", oldest_movie[0], "released in", oldest_movie[1])

# all movie titles
print("All movie titles:")
for movie in movies:
    print("-", movie[0])

# Try to change a year
try:
    movies[0][1] = 2020
except TypeError as error:
    print("\nError occurred when trying to change a year:")
    print(error)
`````

### **Activity 5: Unique Hobbies Set** 


Two friends share their hobbies:
alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies = {"gaming", "hiking", "photography", "music"}

Find and print:
- Hobbies both enjoy (intersection)
- Hobbies only Alex enjoys
- All unique hobbies combined

---

`````python
#hobbies set 
alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies = {"gaming", "hiking", "photography", "music"}

# Hobbies both enjoy
common_hobbies = alex_hobbies & sam_hobbies
print("Hobbies both enjoy:", common_hobbies)

# Hobbies only Alex enjoys
alex_unique = alex_hobbies - sam_hobbies
print("Hobbies only Alex enjoys:", alex_unique)

# All unique hobbies combined
all_hobbies = alex_hobbies | sam_hobbies
print("All unique hobbies combined:", all_hobbies)
`````


### **Activity 6: Simple Grade Book Dictionary** 

Create a dictionary for 3 students and their scores:
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

Perform these tasks:
- Add a new student "Diana" with score 88
- Update Bob's score to 95
- Print all students with scores above 80
- Calculate and print the average score

---

`````python
#book dictionary 
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

# new student 
grades["Diana"] = 88

# Update Bob's score 
grades["Bob"] = 95

# all students with scores above 80
print("Students with scores above 80:")
for student, score in grades.items():
    if score > 80:
        print(f"{student}: {score}")

# Calculate and print the average score
average_score = sum(grades.values()) / len(grades)
print("\nAverage score:", average_score)
`````
