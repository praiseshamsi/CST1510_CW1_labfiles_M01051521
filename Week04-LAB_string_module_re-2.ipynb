3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
#Password to ASCII Converter
def password_to_ascii(password):
    ascii_string = ''
    for char in password:
        ascii_string += str(ord(char))
    return ascii_string
# Main function 
def main():
    password = input("Enter your password: ")
    converted = password_to_ascii(password)
    
    print(f"Your ASCII-encoded password is: {converted}")
    print(f"Original Password : {password}")
    print(f"Your ASSCII encoded passward is: {converted}")

if __name__ == "__main__":
    main()
````

4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
# Password Validator 
def check_password(password):
    if ' ' in password:
        print("Error: Password should not contain spaces.")
        return False 
    # Check each condition
    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in '@#$%' for char in password)

    print(f"Contains uppercase letter: {has_upper}")
    print(f"Contains lowercase letter: {has_lower}")
    print(f"Contains digit: {has_digit}")
    print(f"Contains special character (@#$%): {has_special}")

    valid_password = has_upper and has_lower and has_digit and has_special

    print(f"\nPassword compliance: {valid_password}")
    return valid_password

def main():
    print("Password Validator")
    while True:
        password = input("Enter your password (q to quit): ")
        if password.lower() == "q":
            break
    check_password(password)
    print("-" *10)

if __name__ == "__main__":
    main()
````

10. Create a function (s) counting all lower case, upper case, digits, and special symbols from a given string and print them out.
   
```python

   example_password = "P@#yn26at^&i5ve"
```

````python
def count_characters(s):
    # Counters
    lower = 0
    upper = 0
    digits = 0
    special = 0

    # Go through each character
    for char in s:
        if char.islower():      # check lowercase
            lower += 1
        elif char.isupper():    # check uppercase
            upper += 1
        elif char.isdigit():    # check digit
            digits += 1
        else:                   # everything else is special
            special += 1

    # Print results
    print("Lowercase letters:", lower)
    print("Uppercase letters:", upper)
    print("Digits:", digits)
    print("Special symbols:", special)

example_password = "P@#yn26at^&i5ve"
count_characters(example_password)
````

13. Secret Word Generator with Randomization: Ask the user for their first name and surname. Create a function that generates a special secret word that:
Uses the first 3 letters of the first name and last 2 letters of the surname.
Randomly appends a special character (e.g., @, #, $, etc.) at the end of the secret word.

```python
special_chars = "@#$%^&*!"
```

````python
import random

# Function to create the secret word
def generate_secret_word(first, last):
    # Take first 3 letters of first name
    part1 = first[:3]

    # Take last 2 letters of surname
    part2 = last[-2:]

    # List of special characters
    special_chars = "@#$%^&*!"

    # Randomly pick one special character
    symbol = random.choice(special_chars)

    # add and return secret word
    return part1 + part2 + symbol


# user input
first_name = input("Enter your first name: ")
surname = input("Enter your surname: ")

# Generate and print secret word
secret = generate_secret_word(first_name, surname)
print("Your secret word is:", secret)
````

15. (Binary to decimal) Write a function that parses a binary number as a string and convert it into a decimal integer. Use the function header: def binaryToDecimal(binaryString): For example, binary string 10001 is 17 (1 24 + 0
23 + 0 22 + 0 21 + 1 * 20 = 17)

So, binaryToDecimal("10001") returns 17.

Write a test program that prompts the user to enter a binary string and displays the corresponding decimal integer value.

````python
# Function to convert binary string to decimal
def binaryToDecimal(binaryString):
    decimal = 0
    power = len(binaryString) - 1   # start from highest power

    for digit in binaryString:
        decimal += int(digit) * (2 ** power)
        power -= 1

    return decimal

binary_input = input("Enter a binary string: ")
print("Decimal value:", binaryToDecimal(binary_input))
````

16. (Binary to hex) Write a function that parses a binary number and converts itinto a hex number. The function header is: def binaryToHex(binaryValue):
Write a test program that prompts the user to enter a binary number and displays the corresponding hexadecimal value.



````python
# Function to convert binary to hex
def binaryToHex(binaryValue):
    # convert binary to decimal
    decimal = 0
    power = len(binaryValue) - 1

    for digit in binaryValue:
        decimal += int(digit) * (2 ** power)
        power -= 1

    # convert decimal to hex 
    return hex(decimal)[2:].upper()   # remove "0x" and make uppercase

binary_input = input("Enter a binary number: ")
print("Hex value:", binaryToHex(binary_input))
````

17. (Decimal to binary) Write a function that converts a decimal number into a binary number as a string. Use the function header: def decimalToBinary(value):
Write a test program that prompts the user to enter a decimal integer value and displays the corresponding binary value.

````python
# Function to convert decimal to binary
def decimalToBinary(value):
    binary = ""                 # empty string   

    # If the value is 0
    if value == 0:
        return "0"

    # Convert using repeated division by 2
    while value > 0:
        remainder = value % 2
        binary = str(remainder) + binary
        value //= 2

    return binary


# Test Program 
num = int(input("Enter a decimal number: "))
print("Binary value:", decimalToBinary(num))
````
