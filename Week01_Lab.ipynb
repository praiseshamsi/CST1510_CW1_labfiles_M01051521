### Typical 1. Total Time to Watch a Series

Compute total viewing time (no breaks) for a series:
- 5 seasons, 10 episodes per season
- Season 1: 45 minutes 35 seconds per episode
- Seasons 2–5: 35 minutes 15 seconds per episode
Display the total as hours, minutes, and seconds.

**Example Output:**
```
Total time: 39h 30m 0s
```


````python
# Episode durations in seconds
season1_episode = 45*60 + 35       # 45 minutes 35 seconds
other_seasons_episode = 35*60 + 15 # 35 minutes 15 seconds

# Episodes per season
episodes_per_season = 10

# Total seconds for season 1
season1_total = season1_episode * episodes_per_season

# Total seconds for seasons 2–5 (4 seasons)
other_seasons_total = other_seasons_episode * episodes_per_season * 4

# Total seconds for entire series
total_seconds = season1_total + other_seasons_total

# Convert total seconds to hours, minutes, seconds
hours = total_seconds // 3600
remaining = total_seconds % 3600
minutes = remaining // 60
seconds = remaining % 60

# Output
print(f"Total time: {hours}h {minutes}m {seconds}s")
````

### Typical 2. Savings Projection (3 Years)

Compute final savings after 3 years given:
- Save £10 every week for 3 years
- Year 1 bonus: +£20 each week
- Every 2 weeks: −£0.50 (sister)
Assume 52 weeks per year; use arithmetic and augmented assignment; no conditionals.

**Example Output:**
```
Final savings after 3 years: £4689.50
```


````python
# Weekly amounts
weekly_save = 10.0
year1_bonus = 20.0
biweekly_loss = 0.50

# Time periods
weeks_per_year = 52
years = 3

# Total weeks
total_weeks = years * weeks_per_year
weeks_year1 = weeks_per_year
weeks_other = total_weeks - weeks_year1

# Start savings
total = 0.0

# Add Year 1 savings
total += (weekly_save + year1_bonus) * weeks_year1

# Add Years 2 and 3 savings
total += weekly_save * weeks_other

# Subtract sister’s biweekly losses
biweekly_events = total_weeks // 2
total -= biweekly_loss * biweekly_events

# Final result
print(f"Final savings after 3 years: £{total:.2f}")
````

### Stretch 4. Payroll Statement

Ask for: employee name, hours worked, hourly rate, federal tax rate, state tax rate.
Print a formatted payroll slip with gross pay, each deduction, and net pay (2 d.p.).

**Example Output:**
```
Employee: Smith
Gross pay: £97.50
Federal withholding: £19.50
State withholding: £8.78
Net pay: £69.22
```


````python
# NAME: PRAISE SHAMSI AYEBAZIBWE 
# MISIS: M01051521


# user inputs
employee_name = input("Enter employee name: ")
hours_worked = float(input("Enter hours worked: "))
hourly_rate = float(input("Enter hourly pay rate: "))
federal_tax_rate = float(input("Enter federal tax rate (e.g. 0.2): "))
state_tax_rate = float(input("Enter state tax rate (e.g. 0.09): "))

# Calculation of gross pay
gross_pay = hours_worked * hourly_rate

# Calculation of deductions
federal_withholding = gross_pay * federal_tax_rate
state_withholding = gross_pay * state_tax_rate
total_deductions = federal_withholding + state_withholding

# Calculate net pay
net_pay = gross_pay - total_deductions

# print the formatted payroll slip
print("\nEmployee:", employee_name)
print(f"Gross pay: £{gross_pay:.2f}")
print(f"Federal withholding: £{federal_withholding:.2f}")
print(f"State withholding: £{state_withholding:.2f}")
print(f"Net pay: £{net_pay:.2f}")
````
