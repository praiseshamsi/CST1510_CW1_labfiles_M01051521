4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
# Program that measures Body mass index(BMI)
# User prmpt
user_weight_lbs = int(input("Enter weight in pounds:"))
user_height_inches = int(input("Enter height in inches:"))

kg = user_weight_lbs * 0.45359237
meters = user_height_inches * 0.0254

bmi = kg/(meters*meters)
print(f"Your BMI is: {bmi:.2f}")

# Interpretation for BMI
if bmi < 18.5:
    print("Underweight")
elif bmi <= 25.0:
    print("Normal")
elif bmi <= 30.0:
    print("Overweight")
else:
    print("Obese")
````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
# Marks of students
marks_of_students = [40,45,52,59,60,64,75,88,90,95]

# Variables
total_marks = 0 
passed_students = 0 
failed_students = 0 

# Variable for pass mark
pass_mark = 50

#loop for total mark
for mark in marks_of_students:
    total_marks += mark
    # For passed or failed 
    if mark >= pass_mark:
        passed_students += 1
    else:
        failed_students += 1

#highest mark and lowest 
highest_mark = max(marks_of_students)
lowest_mark = min(marks_of_students)


# Average mark 
number_of_students = len(marks_of_students)

# incase the list is empty
if number_of_students > 0:
    average_mark = total_marks / number_of_students
else:
    average_mark = 0

# Print analysis results
print("STUDENT MARK ANALYSIS")
print(f"Total numer of students: {number_of_students}")
print("-" * 20)


print(f"Average Mark: {average_mark:.2f}")

print(f"Highest Mark: {highest_mark}")
print(f"Lowest Mark: {lowest_mark}")

print("-" * 20)
print(f"Passing Students (Marks >= {pass_mark}): {passed_students}")
print(f"Failing Students (Marks < {pass_mark}): {failed_students}")
````

15. **Find the highest score**

Write a program that prompts the user to enter the number of students and each student’s name and score and display the student name and highest score.


````python
n = int(input("Enter number of students: "))

highest_score = -1
highest_name = ""

for i in range(n):
    name = input("Enter student name: ")
    score = float(input("Enter score: "))

    # Check for the highest score
    if score > highest_score:
        highest_score = score
        highest_name = name

# Print result
print("\nHighest score:", highest_score)
print("Student name:", highest_name)
````

20. (Find the two highest scores) - Write a program that prompts the user to enter the number of students and each student’s score and displays the highest and second highest scores.


````python
# NAME: PRAISE SHAMSI AYEBAZIBWE 
# MISIS: M01051521


#Finding Two Highest scores

# User prompt
num_students = int(input("Enter the number of students: "))

# variables for highest and second highest scores
highest = -float('inf')
second_highest = -float('inf')

# Loop for each student's score
for i in range(num_students):
    score = float(input(f"Enter score for student {i+1}: "))
    
    if score > highest:
        second_highest = highest
        highest = score
    elif score > second_highest:
        second_highest = score

# Print the results
print(f"Highest score: {highest}")
print(f"Second highest score: {second_highest}")
````
